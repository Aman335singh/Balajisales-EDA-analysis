```python
cat_cols=df.select_dtypes(include=['object']).columns
num_cols=df.select_dtypes(include=np.number).columns.tolist()
print('categorical variables')
print(cat_cols)
print('numerical variables')
print(num_cols)
```

```python
for col in num_cols:
    print(col)
    print('Skew: ',round(df[col].skew(),2))
    pt.figure(figsize= (15,3))
    pt.subplot(1,2,1)
    df[col].hist(grid=False)
    pt.ylabel('count')
    pt.subplot(1,2,2)
    sb.boxplot(x=df[col])
    pt.show()
```

```python
def log_tranform(df,col):
    for colname in col:
        if(df[colname]==1.0).all():
            df[colname+ '_log']=np.log(df[colname]+1)
        else:
            df[colname+ '_log']=np.log(df[colname])
    df.info()
```

```python
log_tranform(df,['item_price','quantity','transaction_amount'])
```

```python
sb.distplot(df['transaction_amount_log'], axlabel="transaction_amount_log");
```

```python
sb.distplot(df['item_price_log'], axlabel="item_price_log");
```

```python
sb.distplot(df['quantity_log'], axlabel="quantity_log");
```

```python
pt.figure(figsize=(13,17))
sb.pairplot(data=df.drop(['item_price','quantity','transaction_amount'],axis=1)) #for numerical variables
```
